```python
class Game:
    def __init__(self, board = [[0,0,0],[0,0,0],[0,0,0]]):
        self.board = board
        
    def draw_board(self):
        match = ["   |", " X |", " O |"]
        print(" --- --- --- ")
        for i in range(3):
            print("|", end = "")
            for j in range(3):
                number = self.board[i][j]
                print(match[number], end = "")
            print("\n --- --- --- ")
    
    def player_move(self, p_row, p_column, player):
        if self.board[p_row-1][p_column-1] != 0:
            print("The location is occupied!")
            step = input("Player %s:" % (player)).split()
            self.player_move(int(step[0]),int(step[1]), player)
        else:
            self.board[p_row-1][p_column-1] = player
            self.draw_board()
            
    def check_full(self):
        full = True
        for i in self.board:
            if 0 in i:
                full = False
        return full
    
    def check_board(self):
        player_one_win = False
        player_two_win = False
        if (self.board[0][0] == self.board[2][2] == self.board[1][1])\
        or (self.board[0][2] == self.board[2][0] == self.board[1][1]):
            if self.board[1][1] == 1:
                player_one_win = True
            elif self.board[1][1] == 2:
                player_two_win = True
                
        for x in range(3):
            #row
            row = set([self.board[x][0], self.board[x][1], self.board[x][2]])
            if len(row) == 1 and 0 not in row:
                if 1 in row:
                    player_one_win = True
                elif 2 in row:
                    player_two_win = True
            #column
            column = set([self.board[0][x], self.board[1][x], self.board[2][x]])
            if len(column) == 1 and 0 not in column:
                if 1 in column:
                    player_one_win = True
                elif 2 in column:
                    player_two_win = True
        if player_one_win == True:
            return "One"
        elif player_two_win == True:
            return "Two"
        else:
            if self.check_full() == False:
                return "Neither"
            else:
                return "Draw"
            
    def start_up(self):
        self.board = [[0,0,0],[0,0,0],[0,0,0]]
        self.draw_board()
        player = 1
        while True:
            step = input("Player %s:" % (player)).split()
            self.player_move(int(step[0]),int(step[1]), player)
            if self.check_board() == "One":
                print("Player one wins!")
                break
            elif self.check_board() == "Two":
                print("Player two wins!")
                break
            elif self.check_board() == "Draw":
                print("Draw!")
                break
            player = player%2+1
    
a = Game()
a.start_up()
```
